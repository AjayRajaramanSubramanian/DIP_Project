```python
import matplotlib.pyplot as plt
```

```python
array1=[0.65,0.6,0.44,0.32,0.26]
array2=[0.71,0.83,0.6,0.64,0.56]
```

```python
# Plotting
plt.plot(array1, label='Before Denoising')
plt.plot(array2, label='After Denoising')

# Adding labels and title
plt.xlabel('Level of Noise')
plt.ylabel('Confidence Score')
plt.title('Object Detection Confidence Scores Before and After Denoising')

# Adding legends
plt.legend()

# Show the plot
plt.show()
```
